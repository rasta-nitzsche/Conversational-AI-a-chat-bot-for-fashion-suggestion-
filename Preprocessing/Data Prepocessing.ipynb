# Removing similar items (by comparing the boundaries)

We keep the attribute with the lowest probability (because it is the most detailed) : for example shoe has more proba than sneaker but we would like to keep sneaker. One other way to do it is to keep both but with a pre defined structure

```python
def distance(a,attr1, attr2):
    
    x1,y1,x2,y2 = a[attr1]
    a1,b1,a2,b2 = a[attr2]

    center1 = np.array([ x1 + (x2-x1)/2 , y1 + (y2-y1)/2])
    center2 = np.array([ a1 + (a2-a1)/2 , b1 + (b2-b1)/2])

    dist = np.linalg.norm(center1 - center2)
        
    return dist
```

```python
def size(a,attr):
    x1,y1,x2,y2 = a[attr]
    return ( x2-x1 + y2-y1 ) / 2
```

```python
def remove_similar_items(b): 
    
    final = b['boudaries(X,y,Width,Height)'].copy()
    deleted = []
    
    
    # Items which centers are really close (less than 30 pixels) are the same cause VinVl detected them twice
    #  We will keep items with lowest proba because they the more detailes (shoe vs sneaker)
    
    for k in b['boudaries(X,y,Width,Height)'].keys():
        for v in b['boudaries(X,y,Width,Height)'].keys():
            if (k != v) and (distance(b['boudaries(X,y,Width,Height)'], k,v)<30):

                ak = eval(b['Attributes'])[k.split()[-1]]['pred_score']
                av = eval(b['Attributes'])[v.split()[-1]]['pred_score']
                if av <= ak and k not in deleted: 
                    del final[k]
                    deleted.append(k)
                if av > ak and v not in deleted:
                    del final[v]
                    deleted.append(v)
                                
#                 print(k,v)
                    
    return final
```
