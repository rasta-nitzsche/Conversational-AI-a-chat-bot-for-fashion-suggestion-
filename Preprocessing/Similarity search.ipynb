```python
def similar_items(attr):
    similar_items=[]
    results = model_SIMCSE.search(attr,top_k=10,threshold=0.779)
    for i in range(len(results)):
        similar_items.append(results[i][0])
    return similar_items
```

```python
from nltk.stem import WordNetLemmatizer
  
lemmatizer = WordNetLemmatizer()
```

```python
def matrix_search(attr,k):
    match = []
    i = 0
    append = True
    
    for a in matrix['bigram'] :
        if attr in a :
            a = tuple(a[1:-1].replace('\'','').split(", "))

            if attr in a[0] :
                wrd = a[1]
            else:
                wrd = a[0]

            if (not wrd in match) and (not attr in wrd): 
                    match.append(wrd)
                    i = i + 1
            if i == k+3:
                break
                
#     for el1 in match:
#         for el2 in match:
#             if el1 != el2 :
#                 if model_SIMCSE.similarity(el1,el2) > 0.7:
#                     match.remove(el2)
    return match,i
```

```python
attr = "blue jeans" 
" ".join([lemmatizer.lemmatize(i) for i in attr.split()])
```

```python
def garment_matching(attr,k):           # Returns k best matches to the given attribute
    
    attr = " ".join([lemmatizer.lemmatize(i) for i in attr.split()])
    i= 0
    match = []

    if attr in items:
        print("Found")
        match,i = matrix_search(attr,k)

    else :
        print("not found")
        similar = similar_items(attr)
        stop = False
        ind = 0
        while (not stop):
            if similar[ind] in items:
                match,i = matrix_search(similar[ind],k)
                if (i>0):
                    stop = True  
            ind = ind + 1

    if (i==0):
        print('This attribute was not found for the garment matching!')
        return None
        
    return match
```

```python
def garment_advice(attr1 , attr2,k=10):
    match = []
    i = 0
    match = garment_matching(attr1,k)
    
    attr2 = " ".join([lemmatizer.lemmatize(i) for i in attr2.split()])
    
    if attr2 in match:
        return True
    else:
        return False
```

```python
garment_matching("blue jean",5)
```

```python
garment_advice("blue jeans" , "white shirt",k=10)
```
